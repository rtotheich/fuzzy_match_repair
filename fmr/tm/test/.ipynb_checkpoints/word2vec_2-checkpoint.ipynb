```python
import gensim
from gensim.models import Word2Vec 
from gensim.models import KeyedVectors
import pandas as pd
from nltk.tokenize import RegexpTokenizer

with open('./test', 'r') as fd:
    sentences = fd.readlines()

# tokenize
tokenizer = RegexpTokenizer(r'\w+')
sentences_tokenized = [w.lower() for w in sentences]
sentences_tokenized = [tokenizer.tokenize(i) for i in sentences_tokenized]

model = Word2Vec("/home/yue.r/gensim-data/glove-wiki-gigaword-100/glove-wiki-gigaword-100.txt")
model.train(sentences_tokenized, total_examples=len(sentences_tokenized), epochs=3)
```

```python
print(model.wv.most_similar('hello', topn=10))
```
