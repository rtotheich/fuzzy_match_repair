```python
import os

from dotenv import load_dotenv, dotenv_values

load_dotenv()
```

```python
from openai import OpenAI

client = OpenAI()
```

```python
with open("context_and_masks/60_69_fms.en.txt") as fd:
    lines_60_69 = [line.strip('\n').replace('[MASK]', '?') for line in fd.readlines()]

print('\n'.join(lines_60_69[:40]))
```

```python
from tqdm.auto import tqdm

predictions = ""

for i in tqdm(range(0, len(lines_60_69))):
    completion = client.chat.completions.create(
        model="gpt-4-turbo",
        messages=[
            {"role": "system", "content": f"You are an expert lexicographer and natural language processing assistant.\
             Additionally, you are highly specialized in parliamentary proceedings. Given a trigram I provide with a \
             '?' character in the center word, I need you to predict the '?' character with the most likely single-word token. \
             Please return one predicted token without any text except the predicted token in your response. Do not provide the \
             surrounding text or any additional information. Do not include the text 'predicting', 'predict', 'prediction', 'predicted' \
             'the predicted token is' or 'The predicted token is' in your response. \
             Do not include any extra characters such as apostrophes, commas, colons, or semicolons in your response. Do not include \
             any newline characters in your response."},
            {"role": "user", "content": f"Here is the trigram:\n{lines_60_69[i]}"}
        ]
    )
    while '\n' in completion.choices[0].message.content or \
        'predicted' in completion.choices[0].message.content or \
        ":" in completion.choices[0].message.content or \
        '"' in completion.choices[0].message.content or \
        '{' in completion.choices[0].message.content or \
        len(completion.choices[0].message.content.split()) > 1 or \
        completion.choices[0].message.content == 'predict':
        completion = client.chat.completions.create(
            model="gpt-4-turbo",
            messages=[
                {"role": "system", "content": f"You are an expert lexicographer and natural language processing assistant.\
                Additionally, you are highly specialized in parliamentary proceedings. Given a trigram I provide with a \
                '?' character in the center word, I need you to predict the '?' character with the most likely single-word token. \
                Please return one predicted token without any text except the predicted token in your response. Do not provide the \
                surrounding text or any additional information. Do not include the text 'predicting', 'predict', 'prediction', 'predicted' \
                'the predicted token is' or 'The predicted token is' in your response. \
                Do not include any extra characters such as apostrophes, commas, colons, or semicolons in your response. Do not include \
                any newline characters in your response."},
                {"role": "user", "content": f"Here is the trigram:\n{lines_60_69[i]}"}
            ]
        )

    predictions += completion.choices[0].message.content + '\n'

print(predictions)

with open('predictions/60_69.txt', 'w') as fd:
    fd.write(predictions)
```
